```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
from matplotlib import rcParams
```

```python
incidents = pd.read_csv('../JSONS/Portugal.csv', sep = ',', encoding = 'latin-1')
incidents.head(4)
```

```python
incidents = incidents.drop_duplicates(subset = 'ID', keep = "last")
incidents
```

```python
incidents = incidents.drop(columns = ['Descricoes', 'Coordenadas', 'Categoria', 
                                      'Data_Fim', 'Hora_Fim', 'ID', 'Estradas', 'Versao'])
incidents
```

```python
incidents = incidents.rename(columns = {'Numero_Estradas': 'Number roads affected', 
                                        'Comprimento': 'Length', 'Cidade': 'City'})
```

```python
incidents['City'].value_counts()
```

```python
import datetime

def concat_str(date, hour):
    s = str(date) + ' ' + str(hour)
    s = datetime.datetime.strptime(s, '%Y-%m-%d %H')
    return s.strftime("%Y%m%d%H")
```

```python
incidents['Date'] = incidents.apply(lambda x: concat_str(x['Data_Inicio'], x['Hora_Inicio']), axis = 1)
incidents = incidents.drop(columns = ['Data_Inicio', 'Hora_Inicio'])
```

```python
incidents['Number of incidents'] = 1
incidents = incidents.groupby(['Date']).sum()
incidents
```

```python
incidents.head(20)
```

```python
incidents.reset_index(inplace = True)
l = sorted(list(incidents['Date']))
```

```python
incidents
```

```python
import re

def get_hours_by_day(dates, l):
    
    day = str(dates)[0:8]
    
    number_hours = len(re.findall(day + '[0-9]+\'', str(l)))

    if number_hours < 24:
        return 0
    else:
        return 1
   
incidents['Day has 24 hours'] = incidents.apply(lambda row: get_hours_by_day(row['Date'], l), 
                                                axis = 1)
```

```python
incidents['Day has 24 hours'].value_counts()
```

```python
incidents = incidents[incidents['Day has 24 hours'] == 1]
incidents = incidents.drop(columns = 'Day has 24 hours')
```

```python
incidents = incidents.set_index('Date')
```

```python
corr = incidents.corr("spearman")

print('Spearman Correlation')
print(corr) 

plt.figure(figsize = (10, 8))
sns_plot = sns.heatmap(corr[(corr >= 0.001) | (corr <= -0.001)], cmap = 'magma', 
                       vmax = 1, vmin = -1.0, linewidths = 0.2, annot = True, 
                       annot_kws = {"size": 8}, square = True)
```
